```python
import cv2 
```

```python
img = cv2.imread("elephant.jpg")
print(img.shape)
img = cv2.resize(img, (700,700))
print(img.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(0, figsize = (10,10))
plt.imshow(new_img)
plt.axis("off")
plt.show()
```

```python
all_pixels = new_img.reshape((-1,3))
```

```python
print(all_pixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 20)
```

```python
kmeans.fit(all_pixels)
```

```python
import numpy as np
```

```python
np.unique(kmeans.labels_, return_counts = True)
```

```python
kmeans.cluster_centers_
```

```python
centers = np.array(kmeans.cluster_centers_, dtype = "uint8")
```

```python
print(centers)
```

```python
plt.figure(0, figsize = (20,10))
i = 1
colors = []
for each_col in centers:
    plt.subplot(1,20,i)
    plt.axis("off")
    i+=1
    colors.append(each_col)
    a = np.zeros((100,100,3), dtype = "uint8")
    a[:,:,:] = each_col
    plt.imshow(a)
plt.show()
    
```

```python
seg_img = np.zeros((700*700,3), dtype = "uint8")
```

```python
np.unique(kmeans.labels_, return_counts = True)
```

```python
print(colors)
```

```python
for ix in range(seg_img.shape[0]):
    seg_img[ix] = colors[kmeans.labels_[ix]]
```

```python
seg_img = seg_img.reshape((700,700,3))
print(seg_img.shape)
```

```python
plt.figure(0, figsize = (10,10))
plt.imshow(seg_img)
plt.show()
```
